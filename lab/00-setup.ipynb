## 1. Imports and configuration

We read the endpoint, key, and model names from environment variables
(set in your `.env` file, or provided by the lab environment).


```python
from cohere_helpers import embed_texts, rerank, cosine_search, ENDPOINT, EMBED_MODEL, RERANK_MODEL
import json

print('Endpoint:', ENDPOINT)
print('Embed model:', EMBED_MODEL)
print('Rerank model:', RERANK_MODEL)
```

## 2. Load the support corpus

The corpus is a set of ~150 customer support documents (FAQs, articles, tickets)
for a fictional B2B software platform. This is what we will search.


```python
with open('support_corpus.json', 'r', encoding='utf-8') as f:
    corpus = json.load(f)

print(f'Loaded {len(corpus)} documents')
print('Example document:')
print(json.dumps(corpus[0], indent=2))
```

## 3. Quick embed test

Embed a single short text to confirm the Embed-v4 endpoint works.
The output is a numeric vector (1536 dimensions for Embed-v4).


```python
test_vec = embed_texts(['How do I reset my password?'], input_type='search_query')
print('Embedding shape:', test_vec.shape)
print('First 8 values:', test_vec[0][:8])
```

## 4. Quick rerank test

Rerank a few documents against a query to confirm the Rerank-v4 endpoint works.
The result is sorted best-first, with a relevance score per document.


```python
docs = [
    'Reset your password from the account settings page.',
    'Our billing cycle runs monthly.',
    'Deploy a model from the Foundry catalog.',
]
results = rerank('how do I change my password', docs, top_n=3)
for r in results:
    print(f"score={r['relevance_score']:.3f}  doc={docs[r['index']]}")
```
