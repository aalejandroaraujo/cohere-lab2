# Lab 2 - Notebook 1: Embed the Corpus

**Goal:** turn every support document into a vector with Embed-v4, so we can search by meaning.

**Concept:** an embedding maps text into a point in high-dimensional space.
Texts with similar meaning land near each other. That is what lets us search
by *meaning* instead of *keywords*.


## 1. Load corpus and helpers


```python
from cohere_helpers import embed_texts
import json, numpy as np

with open('support_corpus.json', 'r', encoding='utf-8') as f:
    corpus = json.load(f)
print(f'{len(corpus)} documents loaded')
```

## 2. Build the text to embed

We embed the title plus content of each document. Combining them gives the model
more context than either alone.


```python
doc_texts = [f"{d['title']}. {d['content']}" for d in corpus]
print('Example text to embed:')
print(doc_texts[0])
```

## 3. Embed the whole corpus

We send the documents to Embed-v4 with `input_type='search_document'`.
(Queries later use `input_type='search_query'`; Cohere optimizes each differently.)

The embedding service accepts a limited number of texts per request, so the helper
automatically sends them in batches. This takes about 20 to 30 seconds for ~150 documents.


```python
doc_vectors = embed_texts(doc_texts, input_type='search_document')
print('Corpus embedded.')
print('Matrix shape:', doc_vectors.shape, '  (documents x dimensions)')
```

## 4. Save the vectors

We store the vectors on disk so the next notebook can search them without re-embedding.
The file `doc_vectors.npy` is saved in this folder, on the machine running the notebook.


```python
np.save('doc_vectors.npy', doc_vectors)
print('Saved doc_vectors.npy')
```
